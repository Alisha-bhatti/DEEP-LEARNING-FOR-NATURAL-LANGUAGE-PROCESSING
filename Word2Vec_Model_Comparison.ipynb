```python
import gensim
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
from gensim.test.utils import common_texts
import numpy as np
```

```python
text_corpus = [
    "king queen man woman prince princess royal palace",
    "paris france berlin germany madrid spain rome italy",
    "apple orange banana fruit mango grape",
    "dog cat animal pet puppy kitten",
    "car bus train vehicle transport road",
    "father mother son daughter family",
    "man king royal crown throne",
    "woman queen royal crown throne"
]
```

```python
sentences = [simple_preprocess(sentence) for sentence in text_corpus]
```

```python
sentences
```

```python
cbow_model = Word2Vec(
    sentences,
    vector_size=100,
    window=3,
    min_count=1,
    workers=4,
    sg=0,
    epochs=200
)
```

```python
skipgram_model = Word2Vec(
    sentences,
    vector_size=100,
    window=3,
    min_count=1,
    workers=4,
    sg=1,
    epochs=200
)
cbow_model.wv.similarity("king", "queen")
```

```python
model=cbow_model.wv.most_similar(positive=["king",  "woman"],negative=["man"],topn=3)
```

```python
# Evaluation Function
def evaluate_model(model):
  correct = 0
  result = model.wv.most_similar(positive=["king","woman"],negative=["man"],topn=3)
  if result[0][0] == "queen":
      correct += 1
  return correct
print("CBOW Analogy Score:", evaluate_model(cbow_model))
print("Skip-Gram Analogy Score:", evaluate_model(skipgram_model))
```
